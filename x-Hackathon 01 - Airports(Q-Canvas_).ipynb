##Join 'plane-data.csv' to 'small.csv' to 'airport-codes-na.txt'

###1) Remove any NULL rows from 'plane-data.csv'
###2) Manually build Schema (Column names and Data types) for 'small.csv'
###3) Remove any non-USA Countries from 'airport-codes-na.txt'
###4) Write most efficient code for query below:
###  Query: SELECT 'UniqueCarrier', 'manufacturer', 'model', 'Country', 'IATA', count of 'DepDelay'
###  GROUP BY non-Aggregated columns
###  WHERE 'DepDelay' > 180 minutes. ORDER BY 'DepDelay' Descending

```python
# 'plane-data.csv' code here
plane_df = spark.read.load("dbfs:/databricks-datasets/asa/planes/plane-data.csv", 
                       format="csv", header=True, schema='tailnum string,type string,manufacturer string,issue_date string,model string,status string,aircraft_type string,engine_type string,year string')
```

```python
plane_df = plane_df.dropna()
```

```python
plane_df
```

```python
# 'small.csv' code here
small_df = spark.read.load("dbfs:/databricks-datasets/asa/small/small.csv", 
                       format="csv", header=True, schema='Year string , Month string , DayofMonth string , DayOfWeek string , DepTime string , CRSDepTime string , ArrTime string , CRSArrTime string , UniqueCarrier string , FlightNum string , TailNum string , ActualElapsedTime string , CRSElapsedTime string , AirTime string , ArrDelay string , DepDelay int , Origin string , Dest string , Distance string , TaxiIn string , TaxiOut string , Cancelled string , CancellationCode string , Diverted string , CarrierDelay string , WeatherDelay string , NASDelay string , SecurityDelay string , LateAircraftDelay string')
```

```python
# 'airport-codes-na.txt' here
airport_df = spark.read.option("delimiter", "\t").load("dbfs:/databricks-datasets/flights/airport-codes-na.txt", 
                       format="csv", header=True, schema='City string, State string, Country string,IATA string')
```

```python
airport_df = airport_df.filter(airport_df.Country == "USA")
airport_df.show()
```

```python
# 3-way Join code here with answer 
small_df = small_df.select("UniqueCarrier","TailNum","Origin","Dest","DepDelay")
plane_df = plane_df.select("manufacturer","tailnum","model")
airport_df = airport_df.select("IATA","Country")
joined_df = small_df.filter(small_df.DepDelay >180).join(plane_df,small_df.TailNum ==  plane_df.tailnum,"left")
joined_df = joined_df.join(airport_df,(joined_df.Origin == airport_df.IATA) | (joined_df.Dest == airport_df.IATA),"left")
# Using (joined_df.Origin == airports_df.IATA) | (joined_df.Dest == airports_df.IATA) as joined condition because it wasn't clearly mentioned which columns to join on and either origin or destination could be used as IATA. 
# This would cause the run time to increase but we thought it might be consider than use either Origin or destination
```

```python
import pyspark.sql.functions as F
joined_df=joined_df.select('UniqueCarrier', 'manufacturer', 'model', 'Country', 'IATA','DepDelay')
joined_df = joined_df.groupBy('UniqueCarrier', 'manufacturer', 'model', 'Country', 'IATA').count()
joined_df = joined_df.withColumn("DepDelayCount",F.col("count")).drop("count")
joined_df = joined_df.orderBy("DepDelayCount", ascending=False)
joined_df.show()
```

```python
# doing the same 3 way join using inner join as the question does not mention which join to use.
joined_df_inner = small_df.filter(small_df.DepDelay >180).join(plane_df,small_df.TailNum ==  plane_df.tailnum)
joined_df_inner = joined_df_inner.join(airport_df,(joined_df_inner.Origin == airport_df.IATA) | (joined_df_inner.Dest == airport_df.IATA))
# Using (joined_df.Origin == airports_df.IATA) | (joined_df.Dest == airports_df.IATA) as joined condition because it wasn't clearly mentioned which columns to join on and either origin or destination could be used as IATA. 
# This would cause the run time to increase but we thought it might be consider than use either Origin or destination
```

```python
joined_df_inner=joined_df_inner.select('UniqueCarrier', 'manufacturer', 'model', 'Country', 'IATA','DepDelay')
joined_df_inner = joined_df_inner.groupBy('UniqueCarrier', 'manufacturer', 'model', 'Country', 'IATA').count()
joined_df_inner = joined_df_inner.withColumn("DepDelayCount",F.col("count")).drop("count")
joined_df_inner = joined_df_inner.orderBy("DepDelayCount", ascending=False)
joined_df_inner.show()
```
